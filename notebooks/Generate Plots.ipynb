# Generate Plots for Paper

```python
import matplotlib.pyplot as plt
import numpy as np

# Positions for 2D visualization
anchor = np.array([0, 0])
positive = np.array([1, 0.5])
negative = np.array([2.2, 1.5])  # farther than margin

margin = 0.5

# Plot points
plt.figure(figsize=(6, 5))
plt.scatter(*anchor, color='blue', label='Anchor', s=100)
plt.scatter(*positive, color='green', label='Positive', s=100)
plt.scatter(*negative, color='red', label='Negative', s=100)

# Draw lines
plt.plot([anchor[0], positive[0]], [anchor[1], positive[1]], 'g--', label='d(a, p)')
plt.plot([anchor[0], negative[0]], [anchor[1], negative[1]], 'r--', label='d(a, n)')

# Annotate margin
midpoint = (anchor + negative) / 2
plt.text(midpoint[0]+0.1, midpoint[1]+0.1, 'margin = 0.5', fontsize=10, color='gray')

# Labels
plt.text(anchor[0]-0.1, anchor[1]-0.2, 'A', fontsize=12)
plt.text(positive[0]-0.1, positive[1]+0.1, 'P', fontsize=12)
plt.text(negative[0]-0.1, negative[1]+0.1, 'N', fontsize=12)

plt.legend()
plt.grid(True)
plt.title("Triplet Loss Visualization")
plt.axis('equal')
plt.tight_layout()
plt.savefig("triplet_loss_diagram.png", dpi=300)
plt.show()
```
